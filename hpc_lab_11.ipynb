```python
!pip install mpi4py
from mpi4py import MPI
import numpy as np
import time

def main():
    comm = MPI.COMM_WORLD
    rank = comm.Get_rank()
    size = comm.Get_size()

    if size < 2:
        if rank == 0:
            print("Run with at least two processes")
        return

    N = 10**6


    if rank == 0:
        data = np.arange(N, dtype='i')
        print("Process 0: Starting non-blocking send...")

        start = time.time()

        req = comm.Isend(data, dest=1, tag=0)


        comp = 0
        for i in range(10**6):
            comp += i * i

        print("Process 0: Computation done during send")

        req.wait()
        end = time.time()

        print(f"Process 0: Send completed in {end - start:.4f} sec")


    elif rank == 1:
        data = np.empty(N, dtype='i')

        print("Process 1: Starting non-blocking recv...")
        start = time.time()

        req = comm.Irecv(data, source=0, tag=0)


        comp = 0
        for i in range(10**6):
            comp += i * i

        print("Process 1: Computation done during recv")

        req.wait()
        end = time.time()

        print(f"Process 1: Recv completed in {end - start:.4f} sec")
        print(f"Process 1: First element = {data[0]}, Last element = {data[-1]}")

if __name__ == '__main__':
    main()
```
