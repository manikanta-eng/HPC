```python
!pip install mpi4py
from mpi4py import MPI
import random

def main():
    comm = MPI.COMM_WORLD
    rank = comm.Get_rank()
    size = comm.Get_size()


    if rank == 0:
        temperatures = []
        print("Master collecting temperatures...\n")

        for i in range(1, size):
            temp = comm.recv(source=i, tag=0)
            print(f"Received from Process {i}: {temp:.2f} Degree Celsius")
            temperatures.append(temp)

        if len(temperatures) == 0:
            print("No worker processes found. Run with multiple processes")
            return

        avg_temp = sum(temperatures) / len(temperatures)
        max_temp = max(temperatures)

        print("\n--- Results ---")
        print(f"Average Temperature = {avg_temp:.2f} Degree Celsius")
        print(f"Maximum Temperature = {max_temp:.2f} Degree Celsius")


    else:
        temperature = random.uniform(20.0, 40.0)
        print(f"Process {rank} sending temperature: {temperature:.2f} °C")

        comm.send(temperature, dest=0, tag=0)

if __name__ == "__main__":
    main()
```
