**Section B — Serial Python Job (Batch Job)**

```python

n = 10_000_000
total = 0

for i in range(n):
    total += i * i

print("Sum of squares:", total)
```

**Section C — Interactive Session Python Code**

**Section E — Scratch I/O + Resource Usage**

```python

import os
import time

scratch = os.environ.get("SLURM_TMPDIR", "/tmp")
output_file = os.path.join(scratch, "result.txt")

with open(output_file, "w") as f:
    for i in range(5):
        f.write(f"Step {i}\n")
        time.sleep(1)

print("Output written to:", output_file)
```
