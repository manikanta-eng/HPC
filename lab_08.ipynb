```python
import time
from numba import njit, prange, set_num_threads, get_num_threads

@njit(parallel=True)
def compute_sum(N):
    total = 0.0
    for i in prange(N):
        total += i * 0.000001
    return total

def main():
    N = 10**7

    max_threads = get_num_threads()
    thread_list = [t for t in [1, 2, 4, 8] if t <= max_threads]

    times = {}

    compute_sum(1000)

    for t in thread_list:
        set_num_threads(t)

        start = time.time()
        compute_sum(N)
        end = time.time()

        exec_time = end - start
        times[t] = exec_time

        print(f"Threads = {t}, Time = {exec_time:.4f} sec")

    T1 = times[thread_list[0]]

    print("\nResults:")
    print("Threads\tTime\tSpeedup\tEfficiency")

    for t in thread_list:
        speedup = T1 / times[t]
        efficiency = speedup / t

        print(f"{t}\t{times[t]:.4f}\t{speedup:.2f}\t\t{efficiency:.2f}")

if __name__ == "__main__":
    main()
```
