**ASSIGNMENT 1**

```python
#!pip install mpi4py
from mpi4py import MPI



def main():
    comm = MPI.COMM_WORLD
    rank = comm.Get_rank()
    size = comm.Get_size()

    if size < 2:
        if rank == 0:
            print("Please run with at least two processes")
        return

    if rank == 0:
        message = "Hello from Master (rank 0)"
        print(f"Process {rank} sending message...")
        comm.send(message, dest=1, tag=0)

    elif rank == 1:
        message = comm.recv(source=0, tag=0)
        print(f"Process {rank} received message: {message}")

if __name__ == "__main__":
    main()
```

**ASSIGNMENT 5**

```python
from mpi4py import MPI

comm = MPI.COMM_WORLD
rank = comm.Get_rank()
size = comm.Get_size()

if rank == 0:
    # Master
    for i in range(1, size):
        comm.send(i * 10, dest=i)

    for i in range(1, size):
        result = comm.recv(source=i)
        print(f"Received result from {i}: {result}")

else:
    # Worker
    data = comm.recv(source=0)
    result = data * 2
    comm.send(result, dest=0)
```
